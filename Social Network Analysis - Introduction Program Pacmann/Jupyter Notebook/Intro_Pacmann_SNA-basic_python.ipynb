## Angka

Angka di python bisa berupa `float` atau `desimal` dan `integer` atau `bilangan bulat`, baik `positif` atau `negatif`.

```python
## integer / bilangan bulat positif

x = 9
print(x)
print(type(x))
```

```python
## float / desimal

y = 0.3
print(y)
print(type(y))
```

```python
## integer / bilangan bulat negatif

z = -3
print(z)
print(type(z))
```

```python
## float / bilangan desimal negatif

i = -3.5
print(i)
print(type(i))
```

Operasi matematika seperti penjumlahan, pengurangan, perkalian dan pembagian bisa dilakukan di python seperti berikut:

- penjumlahan: `angka1 + angka2`
- pengurangan: `angka1 - angka2`
- perkalian: `angka1 * angka2`
- pembagian: `angka1 / angka2` (hasilnya float) atau `angka1 // angka2` (hasiln integer)

```python
## Penjumlahan

add_result = x + y
print(add_result)
```

```python
## Pengurangan

sub_result = x - y
print(sub_result)
```

```python
## Perkalian

mul_result = x * z
print(mul_result)
```

```python
## Pembagian yg hasilnya float

div_result = x / z
print(div_result)
print(type(div_result))
```

```python
## Pembagian yg hasilnya integer

div_result2 = 3 // 3
print(div_result2)
print(type(div_result2))
```

## List

List adalah sebuah kumpulan data yang bisa berisi lebih dari satu data dan bisa memiliki tipe data berbeda di dalamnya. Untuk membuat list, kita menggunakan syntax `[]`

```python
## List angka dari 1 sampai 5

angka = [1, 2, 3, 4, 5]
print(angka)
```

```python
## List nama-nama kota

kota = ['Jakarta','Bandung','Medan','Pontianak','Makassar','Jayapura']
print(kota)
```

List bisa berisi data-data yang memiliki tipe data berbeda

```python
ls = [0,-1,2.5,'jakarta']
print(ls)
```

Kita bisa mengambil elemen yg ada di dalam list dengan cara `list_name[index]`

__index selalu mulai dari nol__

```python
print(len('9'))
```

```python
# elemen pertama
print(ls[0])
```

```python
# elemen kedua
print(ls[1])
```

```python
# elemen terakhir
print(ls[-1])
```

Kita bisa menambahkan elemen ke dalam list menggunakan `list_name.append(element)`

```python
ls.append('bandung')
print(ls)
```

## Dictionary

Dictionary adalah tipe data yg berisi kumpulan pasangan data antara `key` dan `value`. Dictionary ditulis dengan syntax `{key:value}`

```python
## Dictionary negara dan ibu kota nya

dct = {'Indonesia':'Jakarta','Malaysia':'Kuala Lumpur','Inggris':'London','Jepang':'Tokyo'}
print(dct)
```

Kita bisa mengakses `value` dari sebuah `key` di dictionary dengan syntax `dictionary_name[key]`

```python
print(dct['Indonesia'])
print(dct['Jepang'])
```

Jika `key` yg dimasukkan tidak ada di dictionary, maka akan muncul error

```python
dct['Singapore'] = 'Singapore'
print(dct['Singapore'])
```

## If Else

If-else atau if-elif-else bisa digunakan jika kita ingin melakukan sesuatu berdasarkan kondisi.

Jika menggunakan `if-else`, kode yg berada di dalam block `if` akan dijalankan jika kondisi `if` terpenuhi. Sementara kode di dalam block `else` akan dijlankan jika kondisi lainnya terpenuhi.

Jika menggunakan `if-elif-else`, kode yg berada di dalam `elif` akan dijalankan ketika kondisi `elif` terpenuhi. Kode di dalam `else` akan dijalankan jika kondisi `if` dan `elif` tidak ada yg terpenuhi.

```python
## if-else

a = 3

if a > 10:
  print('a lebih besar dari 10')
else:
  print('a lebih kecil dari 10')
```

```python
b = -3

if b > 0:
  print('b bilangan positif')
```

```python
## if-elif-else

a = 3

if a < 5:
  print('a lebih kecil dari 5')
elif a < 15:
  print('a lebih kecil dari 15 dan lebih besar dari 5')
else:
  print('a lebih besar dari 15')
```

## Looping

Terdapat 2 jenis looping di python, yaitu for loop dan while loop.

For loop akan melakukan pengulangan sampai elemen terakhir. Sementara while loop akan melakukan pengulangan terus-menerus selama sebuah kondisi terpenuhi.

```python
# looping dengan for untuk angka dari 1 sampai 10

for i in range(1,11):
    print(i)
```

```python
# looping elemen dalam list

kota = ['Jakarta','Bandung','Medan','Pontianak','Makassar','Jayapura','Denpasar']
for elem in kota:
    print(elem)
```

## Menyimpan atau membaca file

Kita bisa membuat atau membaca sebuah file di python

```python
# membuat file dengan extention txt yang berisi kata 'tes'

f = open('test.txt','w')
f.write('tes')
f.close()
```

```python
# membaca file yang sudah dibuat

f = open('test.txt','r')
text = f.read()
print(text)
```

Selain cara di atas, bisa dengan `with open(file,mode)`

```python
with open('test2.txt','w') as f:
    f.write('tes membuat file dengan with')
```

```python
with open('test2.txt','r') as f:
    text = f.read()
    
print(text)
```
