```python
import json
from tweepy import OAuthHandler, API, Cursor
```

Load file json berisi token dan key twitter api

```python
f = open('twitter_token3.json','r')
token = json.load(f)
f.close()
```

File json tersebut berisi

`{'Access Token' : xxxxxxxxxxxx,
'Access Token Secret' : xxxxxxxxxxxx,
'Consumer Key' : xxxxxxxxxxxx,
'Consumer Key Secret' : xxxxxxxxxxxx}`

```python
token.keys()
```

```python
access_token = token['access_token']
access_secret = token['access_secret']
consumer_key = token['consumer_key']
consumer_secret = token['consumer_secret']

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = API(auth, wait_on_rate_limit = True, timeout = 15*60)
```

Mengambil informasi detail sebuah akun

```python
results = api.lookup_users(screen_names=['solehsolihun','ArditErwandha','pandji'])

account_details = []

for result in results:
    account_details.append(result._json)
```

```python
account_details[0]
```

Timeline dari sebuah akun

```python
res_timeline = api.user_timeline(screen_name='motogp',exclude_replies=False,include_rts=True,count=200,tweet_mode='extended')

acc_timeline = []

for t in res_timeline:
    acc_timeline.append(t._json)
```

Query tweet dengan keyword

```python
## mengambil semua tweet yg ada 'pacmannai' nya dari 7 hari terakhir sampai yg terbaru

query = 'pacmannai'

tweets = []
for tweet in api.search(query, lang='id', count=100):
    tweets.append(tweet._json)
```

```python
tweets[0]
```

```python
## Jika terdapat lebbih dari 100 tweet dan kita ingin dapatkan semua tweet tersebut, bisa pakai fungsi/method Cursor dari tweepy

tweets_all = []

## keyword yg akan kita cari dari sebuah tweet
query = 'pacmannai'

count = 1
for tweet_list in Cursor(api.search, query, count=100, tweet_mode='extended',lang='id').pages():
    tweets_all.extend([i._json for i in tweet_list])
    print('request ke:',count)
    count+=1
```

```python
len(tweets_all)
```

```python
tweets_all[0]
```

```python
with open('pacmann.json','w') as f:
    json.dump(tweets_all,f)
```
